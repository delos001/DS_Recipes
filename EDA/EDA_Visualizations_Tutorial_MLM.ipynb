```python
from matplotlib import pyplot
from pandas import read_csv
```

```python
filename = 'pima-indians-diabetes.data.csv'
path = 'D:\\OneDrive - QJA\\My Files\\DataScience\\DataSets'
names = ['preg', 'plas', 'pres', 'skin', 'test', 
         'mass', 'pedi', 'age', 'class']
```

```python
data = read_csv(path + '\\' + filename,
               names = names)
```

###  Univariate Plots

####  Histogram

```python
data.hist()
pyplot.show()

# !!remember: many ML techques assume Gaussian distribution
```

####  Density Plot

```python
# smooth curve drawn through top of each bin

data.plot(kind = 'density', subplots = True,
         layout = (3,3), sharex = False)
pyplot.show()
```

####  Box and Whisker

```python
# Summaryize distribution using median, 25th and 75th
#    percentile, and whiskers = 1.5 x IQR

data.plot(kind = 'box', subplots = True,
         layout = (3,3), sharex = False,
         sharey = False)
pyplot.show()
```

### Multivariate Plots

####  Correlation Matrix Plot

```python
## indicate of how one variable changes relative to the other
import numpy
correlations = data.corr()

fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin = -1, vmax = 1)
fig.colorbar(cax)

# cutomized portion to set ticks and axis names
ticks = numpy.arange(0,9,1)  # array from 0-9 by 1 increments
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)

pyplot.show()
```

#### Scatter Plot Matrix

```python
from pandas.plotting import scatter_matrix


scatter_matrix(data)
pyplot.show()
```
